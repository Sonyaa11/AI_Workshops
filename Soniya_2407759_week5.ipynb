# **• To - Do:4**
We will define a function that:
1. Loads the data and splits it into training and test sets.
2. Prepares the feature matrix (X) and target vector (Y).
3. Defines the weight matrix (W) and initializes the learning rate and number of iterations.
4. Calls the gradient descent function to learn the parameters.
5. Evaluates the model using RMSE and R2
.


```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.preprocessing import StandardScaler # Import StandardScaler for feature scaling


data = pd.read_csv('/content/drive/MyDrive/WEEK5/student.csv')

# Step 1: Load data and split into train and test sets
X = data.drop('Writing', axis=1)
y = data['Writing']

# Split data into training and testing (70% train, 30% test)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Step 2: Feature Scaling using Standardization
scaler = StandardScaler() # Create a StandardScaler object
X_train = scaler.fit_transform(X_train) # Fit and transform on training data
X_test = scaler.transform(X_test) # Transform test data using the same scaler

# Step 3: Initialize weight matrix (W), learning rate, and number of iterations
m, n = X_train.shape  # m is number of training samples, n is number of features
W = np.random.randn(n)  # Random initialization of weights
learning_rate = 0.01
iterations = 1000

# Step 4: Gradient Descent function
def gradient_descent(X, y, W, learning_rate, iterations):
    m = len(y)
    for i in range(iterations):
        predictions = X.dot(W)
        errors = predictions - y
        gradients = (2/m) * X.T.dot(errors)  # Compute gradients
        W -= learning_rate * gradients  # Update weights
    return W

# Train the model using gradient descent
W_optimal = gradient_descent(X_train, y_train, W, learning_rate, iterations)

# Step 5: Evaluate the model

# Make predictions
y_pred = X_test.dot(W_optimal)

# RMSE (Root Mean Squared Error)
rmse = np.sqrt(mean_squared_error(y_test, y_pred))

# R² (R-squared) score
r2 = r2_score(y_test, y_pred)

print(f'Optimal Weights: {W_optimal}')
print(f'RMSE: {rmse}')
print(f'R²: {r2}')
```
